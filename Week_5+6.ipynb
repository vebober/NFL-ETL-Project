```python
Score_Id5 = []
Week5 = []
Away_Team_Name5 = []
Home_Team_Name5 = []
Home_Team_Score5 = []
Away_Team_Score5 = []
Total_Score5 = []
Home_Money_Line5 = []
Away_Money_Line5 = []
Home_Point_Spread5 = []
Away_Point_Spread5 = []
Home_Point_Spread_Payout5 = []
Away_Point_Spread_Payout5 = []
```

```python
Score_Id6 = []
Week6 = []
Away_Team_Name6 = []
Home_Team_Name6 = []
Home_Team_Score6 = []
Away_Team_Score6 = []
Total_Score6 = []
Home_Money_Line6 = []
Away_Money_Line6 = []
Home_Point_Spread6 = []
Away_Point_Spread6 = []
Home_Point_Spread_Payout6 = []
Away_Point_Spread_Payout6 = []
```

```python
for Score_Data in response5:
    Score_Id5.append(Score_Data["ScoreId"])
    Week5.append(Score_Data["Week"])
    Away_Team_Name5.append(Score_Data["AwayTeamName"])
    Home_Team_Name5.append(Score_Data["HomeTeamName"])
    Home_Team_Score5.append(Score_Data["HomeTeamScore"])
    Away_Team_Score5.append(Score_Data["AwayTeamScore"])
    Total_Score5.append(Score_Data["TotalScore"])
    Home_Money_Line5.append(Score_Data["PregameOdds"][0]["HomeMoneyLine"])
    Away_Money_Line5.append(Score_Data["PregameOdds"][0]["AwayMoneyLine"])
    Home_Point_Spread5.append(Score_Data["PregameOdds"][0]["HomePointSpread"])
    Away_Point_Spread5.append(Score_Data["PregameOdds"][0]["AwayPointSpread"])
    Home_Point_Spread_Payout5.append(Score_Data["PregameOdds"][0]["HomePointSpreadPayout"])
    Away_Point_Spread_Payout5.append(Score_Data["PregameOdds"][0]["AwayPointSpreadPayout"])
```

```python
for Score_Data in response6:
    Score_Id6.append(Score_Data["ScoreId"])
    Week6.append(Score_Data["Week"])
    Away_Team_Name6.append(Score_Data["AwayTeamName"])
    Home_Team_Name6.append(Score_Data["HomeTeamName"])
    Home_Team_Score6.append(Score_Data["HomeTeamScore"])
    Away_Team_Score6.append(Score_Data["AwayTeamScore"])
    Total_Score6.append(Score_Data["TotalScore"])
    Home_Money_Line6.append(Score_Data["PregameOdds"][0]["HomeMoneyLine"])
    Away_Money_Line6.append(Score_Data["PregameOdds"][0]["AwayMoneyLine"])
    Home_Point_Spread6.append(Score_Data["PregameOdds"][0]["HomePointSpread"])
    Away_Point_Spread6.append(Score_Data["PregameOdds"][0]["AwayPointSpread"])
    Home_Point_Spread_Payout6.append(Score_Data["PregameOdds"][0]["HomePointSpreadPayout"])
    Away_Point_Spread_Payout6.append(Score_Data["PregameOdds"][0]["AwayPointSpreadPayout"])
```
